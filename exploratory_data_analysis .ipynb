```python
import pandas as pd

# Content_Rating ve Category bazında gruplama
category_installs = df.groupby(['Content_Rating', 'Category'])['Installs'].sum().reset_index()

# İndirme sayılarının toplamını görmek için
category_installs.sort_values(by='Installs', ascending=False, inplace=True)

category_installs
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Barplot ile gösterim
plt.figure(figsize=(15, 7))
plt.xticks(rotation=60,fontsize=11)
sns.barplot(x='Category', y='Installs', hue='Content_Rating', data=category_installs, dodge=True)
plt.title('Content Rating ve Kategorilere Göre İndirme Sayıları')
plt.xlabel('Toplam İndirme Sayısı')
plt.ylabel('Kategori')

plt.tight_layout()
plt.show()
```
